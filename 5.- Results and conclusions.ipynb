# Results and conclusions

```python
#Setup
from __future__ import division, print_function, unicode_literals

# Common imports
import numpy as np
import os
import pandas as pd
# to make this notebook's output stable across runs
np.random.seed(42)
import matplotlib.pyplot as plt
import pathlib
import seaborn as sns
```

```python
#Importing results
path = pathlib.Path.cwd() / 'results' / "results.csv"
results = pd.read_csv(path, index_col= None)


#Including an Accuracy column
results["Accuracy %"] = 100 - results["Error %"]
#Limiting accuracy to 0
results.loc[results["Accuracy %"] < 0 ,  "Accuracy %"] = 0
#Showing results
results
```

```python
#Describing results metrics
results.describe()[["Predicted", "Expected", "Error abs","Error %", "Accuracy %"]]
```

```python
#Plotting Results
plt.style.use('seaborn-dark')
sns.jointplot(x="Expected" , y ="Predicted", data = results, kind= "reg" , palette= "viridis", color="green",  joint_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.1}})
```
